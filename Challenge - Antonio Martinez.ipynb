```python
"""
Damavis Challenge -> Labyrinth
Developed by -> Antonio Martinez
v.1.0
"""
header = """                                                                                            
 _|                  _|                                                  _|      _|        
 _|          _|_|_|  _|_|_|      _|_|    _|  _|_|  _|    _|  _|_|_|    _|_|_|_|  _|_|_|    
 _|        _|    _|  _|    _|  _|_|_|_|  _|_|      _|    _|  _|    _|    _|      _|    _|  
 _|        _|    _|  _|    _|  _|        _|        _|    _|  _|    _|    _|      _|    _|  
 _|_|_|_|    _|_|_|  _|_|_|      _|_|_|  _|          _|_|_|  _|    _|      _|_|  _|    _|  
                                                         _|                                
                                                     _|_|                                 by Antonio Martinez"""

message = "\nInstructions:\nMove the rod (===) until you reach the lower-right corner, try to avoid the obstacles! (#)"

#Show header and wellcome message
print (header,"\n", message)

# Variables:
key = " " #used to store users input
counter = 0 # stores the number of moves

# Labyrinth creation, put here the labyrinths to complete Acceptance tests:
labyrinth = [[".",".",".",".",".",".",".",".","."],
["#",".",".",".","#",".",".",".","."],
[".",".",".",".","#",".",".",".","."],
[".","#",".",".",".",".",".","#","."],
[".","#",".",".",".",".",".","#","."]]

# 1x3 size rod creation:
rod = ['=','=','=']

# Functions:

# Shows the labyrinth after placing the rod
def _showLab():
    print(*labyrinth, sep = "\n", end = "\n\n")

# Increases the counter and show the current number of moves
def _counter():
        global counter
        counter += 1
        print ("Score:",counter)
        
# counter if players fails        
def _counterFail():
        global counter
        counter = -1
        print ("Score:",counter)

# Places the rod in the start position, upper-left corner
def _startRod():
    try:
        for i in range(len(rod)):
            labyrinth[0][i] = rod[i] #places the rod in the initial position (0, 0)
    except IndexError as e:
            print ("Outside the labyrinth.")
    except Exception as e:
            print (type(e).__name__)
    finally:
            _showLab() 
    
# Gets the position of all blocked cells
def _getBlockedPos():
    blocked = []
    try:
        for i in range(len(labyrinth)):
            for j in range (0,9):           
                if labyrinth[i][j] == '#':
                        blocked.append ([i, j])
        return blocked
    except IndexError as e:
            print ("No existing blocked cells.")
    except Exception as e:
            print (type(e).__name__)
            
# Gets goal cell
def win():    
    goal = []
    goal = [(len(labyrinth)-1), (len(labyrinth[len(labyrinth)-1])-1)]
    return goal

# Resets the labyrinth back to the starting point by replacing all "=" blocks with "."
def _resetLab():
    try:
        for i in range(len(labyrinth)):
            for j in range (0,9):           
                if labyrinth[i][j] == '=':
                    labyrinth[i][j] = '.'  
    except IndexError as e:
            print ("Outside the labyrinth.")
    except Exception as e:
            print (type(e).__name__)            

# Returns the current position of the rod
def currentPosition(): 
    try:
        position = []
        for i in range(len(labyrinth)):
            for j in range (0,9):           
                if labyrinth[i][j] == '=':
                    position.append([i,j])
        return position
    except IndexError as e:
        print ("Outside the labyrinth.")
    except Exception as e:
        print (type(e).__name__)

# Public moveRight() function:
def moveRight():
    if (orientationHor()== True):
        _moveRightHor()
        _counter()
    else:
        _moveRightVer()
        _counter()

# private function: Move the rod to the right if its orientation is horizontal, orientation = True
def _moveRightHor():
    position = currentPosition()
    a = position[2][1] + 1
    newPos = [position[2][0], a] #cell to move to
    blockedPos = _getBlockedPos()
    winPos = win()
    if newPos in blockedPos:  #Check if there is a blocked cell
        print("You hit a blocked cell!")
        _counterFail()
        raise Exception("Game Over.")
    elif newPos in winPos:
        print("You won! Congratulations")
        _counter()
        raise Exception()
    else:
        try:
            labyrinth [position[0][0]][position[0][1]] = '.' #delete the first part of the rod
            labyrinth [position[2][0]][a] = '=' #add one segment to the rod in the next position 'right'
        except IndexError as e:
            print ("You got out of the labyrinth.")
            raise Exception("Game Over.")
        except Exception as e:
            print (type(e).__name__)
        finally:
            _showLab()
    
#private function: Move the rod to the right if its orientation is vertical, orientation = False
def _moveRightVer():
    position = currentPosition() 
    for i in range (0,len(position)):
        labyrinth [position[i][0]][position[i][1]] = '.'
        position[i][1] += 1
        newPos = [position[i][0],position[i][1]]
        winPos = win()
        blockedPos = _getBlockedPos()
        if newPos in blockedPos:  #Check if there is a blocked cell
            print ("You hit a blocked cell!")
            _counterFail()
            raise Exception("Game Over.")
        elif newPos in winPos:
            print("You won! Congratulations")
            _counter()
            raise Exception()
        else:
            labyrinth [position[i][0]][position[i][1]] = '='
            _showLab()

# Public function: moves the rod to the left
def moveLeft():
    if (orientationHor()== True):
        _moveLeftHor()
        _counter()
    else:
        _moveLeftVer()
        _counter()
        
# private function Move rod to the left if its orientation is horizontal
def _moveLeftHor():
    position = currentPosition()     
    b = position[0][1] - 1
    assert b != -1, "You got out of the labyrinth."
    newPos = [position[0][0], b]
    winPos = win()
    blockedPos = _getBlockedPos()
    if newPos in blockedPos:
        print("You hit a blocked cell!")
        _counterFail()
        raise Exception("Game Over.")
    elif newPos in winPos:
        print("You won! Congratulations")
        _counter()
        raise Exception()
    else:
        labyrinth [position[0][0]][b] = '=' #add one segment to the rod in the next position 'left'
        labyrinth [position[2][0]] [position[2][1]] = '.' #delete the last part of the rod
        _showLab()

# private function: Move rod to the left if its orientation is vertical
def _moveLeftVer():
    position = currentPosition()  
    for i in range (0,len(position)):
        labyrinth [position[i][0]][position[i][1]] = '.'
        position[i][1] -= 1 
        assert position[i][1] != -1, "You got out of the labyrinth."
        newPos = [position[i][0], position[i][1]]
        winPos = win()
        blockedPos = _getBlockedPos()
        if newPos in blockedPos:
            print ("You hit a blocked cell!")
            _counterFail()
            raise Exception("Game Over.")
        elif newPos in winPos:
            print("You won! Congratulations")
            _counter()
            raise Exception()
        else:
            labyrinth [position[i][0]][position[i][1]] = '='
            _showLab()

# Public function: moves the rod to bottom
def moveBottom():
    if (orientationHor()== True):
        _moveBottomHor()
        _counter()
    else:
        _moveBottomVer()
        _counter()
    
# private function: Move rod to the bottom if the orientation is horizontal
def _moveBottomHor():
    position = currentPosition()
    for i in range (0,len(position)):
        labyrinth [position[i][0]][position[i][1]] = '.'
        position[i][0] += 1
        newPos = [position[i][0], position[i][1]]
        winPos = win()
        blockedPos = _getBlockedPos()
        if newPos in blockedPos:
            print("You hit a blocked cell!")
            _counterFail()
            raise Exception("Game Over.")
        elif newPos in winPos:
            print("You won! Congratulations")
            _counter()
            raise Exception()
        else:
            labyrinth [position[i][0]][position[i][1]] = '='
    _showLab()
    
    
# private function: Move rod to the bottom if the orientation is vertical
def _moveBottomVer():
    position = currentPosition()
    a = position[2][0] + 1
    newPos = [a, position[2][1]]
    winPos = win()
    blockedPos = _getBlockedPos()
    if newPos in blockedPos:
        print ("You hit a blocked cell!")
        _counterFail()
        raise Exception("Game Over.")
    elif newPos in winPos:
        print("You won! Congratulations")
        _counter()
        raise Exception()
    else:
        labyrinth [position[0][0]][position[0][1]] = '.' #delete the first part of the rod
        labyrinth [a][position[2][1]] = '=' #add one segment to the rod in the next position 'right'
    _showLab()

# Public function: moves the rod to bottom
def moveTop():
    if (orientationHor()== True):
        _moveTopHor()
        _counter()
    else:
        _moveTopVer()
        _counter()
    
# Move rod to the top:
def _moveTopHor():
    position = currentPosition()
    for i in range (0,len(position)):
        labyrinth [position[i][0]][position[i][1]] = '.'
        position[i][0] -= 1
        print (position[i][0])
        assert position[i][1] != -1, "You got out of the labyrinth."
        newPos = [position[i][0], position[i][1]]
        winPos = win()
        blockedPos = _getBlockedPos()
        if newPos in blockedPos:
            print ("You hit a blocked cell!")
            _counterFail()
            raise Exception("Game Over.")
        elif newPos in winPos:
            print("You won! Congratulations")
            raise Exception()
        else:
            labyrinth [position[i][0]][position[i][1]] = '='
            _showLab()
    
# private function Move rod to the left if its orientation is horizontal
def _moveTopVer():
    position = currentPosition()     
    b = position[0][0] - 1
    assert b != -1, "You got out of the labyrinth."
    newPos = [b, position[0][1]]
    winPos = win()
    blockedPos = _getBlockedPos()
    if newPos in blockedPos:
        print ("You hit a blocked cell!")
        _counterFail()
        raise Exception("Game Over.")
    elif newPos in winPos:
        print("You won! Congratulations")
        raise Exception()
    else:
        labyrinth [b][position[0][1]] = '='
        labyrinth [position[2][0]] [position[2][1]] = '.' #delete the last part of the rod
        _showLab()

#Spin the rod:
def spin():
    if (orientationHor() == True):
        _spinHorVer()
    else:
        _spinVerHor()

def checkPos(list):
    pass #check before send
        
# Spin rod from horizontal to vertical position:
def _spinHorVer():
        position = currentPosition()  
        labyrinth [position[0][0]] [position[0][1]] = '.' #delete the tail of the rod
        labyrinth [position[2][0]] [position[2][1]] = '.' #delete the head of the rod
        position[1][0] -= 1
        assert position[1][0] != -1, "You got out of the labyrinth."
        newPos1 = [position[1][0], position[1][1]]
        winPos = win()
        blockedPos = _getBlockedPos()
        if newPos1 in blockedPos:
            print ("You hit a blocked cell!")
            _counterFail()
            raise Exception("Game Over.")
        elif newPos1 in winPos:
            print("You won! Congratulations")
            raise Exception()
        else:
            labyrinth [position[1][0]][position[1][1]] = '=' #spin tail
        position[1][0] += 2
        newPos2 = [position[1][0], position[1][1]]
        winPos = win()
        blockedPos2 = _getBlockedPos()
        if newPos2 in blockedPos:
            print ("You hit a blocked cell!")
            _counterFail()
            raise Exception("Game Over.")
        elif newPos2 in winPos:
            print("You won!")
            raise Exception("Congratulations!")
        else:
            labyrinth [position[1][0]][position[1][1]] = '=' #spin head
            _showLab()

# Spin rod from horizontal to vertical position:
def _spinVerHor():
        position = currentPosition() 
        labyrinth [position[0][0]] [position[0][1]] = '.' #delete the tail of the rod
        labyrinth [position[2][0]] [position[2][1]] = '.' #delete the head of the rod  
        position[1][1] -= 1
        assert position[1][1] != -1, "You got out of the labyrinth."
        newPos1 = [position[1][0], position[1][1]]
        blockedPos = _getBlockedPos()
        if newPos1 in blockedPos:
            print ("You hit a blocked cell!")
            _counterFail()
            raise Exception("Game Over.")
        else:
            labyrinth [position[1][0]][position[1][1]] = '=' #spin tail
        position[1][1] += 2
        newPos2 = [position[1][0], position[1][1]]
        if newPos2 in blockedPos:
            print ("You hit a blocked cell!")
            _counterFail()
            raise Exception("Game Over.")
        else:
            labyrinth [position[1][0]][position[1][1]] = '=' #spin head
            _showLab()

# Get horizontal or vertical orientation:
def orientationHor():  #checks if the orientation is horizontal = True or vertical = False
    orientation = True
    position = currentPosition()
    num1= position[0][0] #get first element of position list
    num2= position[1][0] #get second element of position list
    num3= position[2][0] #get third element of position list
    if num1 == num2 == num3:
        orientation = True
        return orientation
        print ("The orientation is horizontal.")
    else:
        orientation = False
        return orientation
        print ("The orientation is vertical.")

# start and Show the initial labyrinth
_startRod()        
        
# Ask the user to write 'wasd' / 'e' / 'start' / 'quit'
while key != "quit":
    key = input("""Please, press 'w/a/s/d' + Enter to move up/left/down/right the rod through the labyrinth.
Press 'e' to spin the rod.
Type 'start' to restart the game.
Type 'quit' to exit the game.""")   
    key = key.lower()
    if key == "w" or key == "a" or key == "s" or key == "d" or key == "e" or key == "start" or key == "quit":
        if key == "w":
            moveTop()
        elif key == "a":
            moveLeft()
        elif key == "s":
            moveBottom()
        elif key == "d":
            moveRight()
        elif key == "e":
            spin()
        elif key == "start":
            _resetLab()
            _startRod()
        else:
            print ("Good job! Your score is:",counter)
            print ("Leaving the game...")
    else:
        print ("Not valid.")


```
